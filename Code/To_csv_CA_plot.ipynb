```python
import pandas as pd
import matplotlib.pyplot as plt

# path of the csv file.
file_path = r"C:\Users\harsh\OneDrive\Desktop\project\Exp_FeSO4_0.5M_with_H2SO4_0.5\CA\chrono_for_30min.csv"

# reading csv file.
df = pd.read_csv(file_path)

# displaying some input to check.
print(df.head())

# Plotting Time vs Current
plt.figure(figsize=(11, 6))
plt.plot(df["Time (s)"], df["Current (A)"], 'b.-')
plt.xlabel("Time (s)")
plt.ylabel("Current (A)")
plt.title("Chronoamperometry: Constant Potential Experiment")
plt.grid(True)
plt.tight_layout()
plt.show()
```
