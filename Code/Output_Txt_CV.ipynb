```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# path file
file_path = r"C:\Users\harsh\OneDrive\Desktop\project\Exp_FeSO4_with_H2SO4\CV\cv_3_25cycles.txt"

# Readinging and skipping the 17 rows.
df = pd.read_csv(file_path, skiprows=17, header=None, sep=r'\s+')

#column names
df.columns = ['Number', 'Time', 'Potential', 'Current']

# Find anodic and cathodic peak currents
apc = df['Current'].max()
cpc = df['Current'].min()

print(f"Anodic Peak Current: {apc:.6e} A")
print(f"Cathodic Peak Current: {cpc:.6e} A")

# Plotting the CV curve
plt.plot(df['Potential'], df['Current'])
plt.xlabel("Potential (V)")
plt.ylabel("Current (A)")
plt.title("Cyclic Voltammetry Curve")
plt.grid(True)
plt.show()
```

```python
# Potential at anodic peak current
apc_potential = df.loc[df['Current'].idxmax(), 'Potential']

# Potential at cathodic peak current
cpc_potential = df.loc[df['Current'].idxmin(), 'Potential']

print(f"Potential at Anodic Peak Current: {apc_potential:.3f} V")
print(f"Potential at Cathodic Peak Current: {cpc_potential:.3f} V")
```
